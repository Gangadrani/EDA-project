# 3.What can we learn from data?(ex:location,price,reviews,etc )

# 8.What is  the room count in overall NYC according to the listing of room types?


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (8, 5)
ax = sns.countplot(y='room_type', hue='neighbourhood_group', data=airbnb, palette='bright')

total = len(airbnb['room_type'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width() / total)
    x = p.get_x() + p.get_width() + 0.02
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y))

plt.title('Count of each room type in NYC')
plt.xlabel('Rooms')
plt.xticks(rotation=90)
plt.ylabel('Room Counts')

plt.show()
```

Manhattan ha more listed properties with Entire home/apt around 27% of total listed properties  followed by Brooklyn with around 19.6%.

Private rooms are more in Broklyn as in 20.7% of the total listed properties followed by Mahattan with 16.3% of them while 6.9% of private rooms are from Queens

We can infier that Brooklyn ,Queens ,Bronx has more private room types while Manhattan has the highest no  of listing in entire home/apt room types
